```python
%pip install seaborn
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, beta

# Generowanie próbek z rozkładu jednorodnego
uniform_samples = np.random.uniform(0, 1, 1000)

# Generowanie próbek z rozkładu normalnego
mean1, std_dev1 = 0, 1
normal_samples1 = np.random.normal(mean1, std_dev1, 1000)

mean2, std_dev2 = 0, 10
normal_samples2 = np.random.normal(mean2, std_dev2, 1000)

# Generowanie próbek z rozkładu beta
beta_samples1 = np.random.beta(1, 1, 1000)
beta_samples2 = np.random.beta(2, 2, 1000)
beta_samples3 = np.random.beta(3, 9, 1000)

# Generowanie próbek z dwuwymiarowego rozkładu normalnego
mean = [0, 0]
covariance_matrix1 = [[1, 0], [0, 1]]
covariance_matrix2 = [[2, 1], [1, 2]]
multivariate_normal_samples1 = np.random.multivariate_normal(mean, covariance_matrix1, 10000)
multivariate_normal_samples2 = np.random.multivariate_normal(mean, covariance_matrix2, 10000)

# Wizualizacja danych
plt.figure(figsize=(20, 10))

plt.subplot(2, 3, 1)
plt.hist(uniform_samples, bins=20, density=True, color='skyblue', edgecolor='black')
plt.title('Uniform')


plt.subplot(2, 3, 2)
plt.hist(normal_samples1, bins=20, density=True, color='salmon', edgecolor='black')
plt.plot(np.linspace(-3, 3, 1000), norm.pdf(np.linspace(-3, 3, 1000), loc=0, scale=1), 'g')
plt.title('Rozkład normalny (mean=0, std_dev=1)')


plt.subplot(2, 3, 3)
plt.hist(normal_samples2, bins=20, density=True, color='lightgreen', edgecolor='black')
plt.plot(np.linspace(-40, 40, 1000), norm.pdf(np.linspace(-40, 40, 1000), loc=0, scale=10), 'g')
plt.title('Rozkład normalny (mean=0, std_dev=10)')


plt.subplot(2, 3, 4)
plt.hist(beta_samples1, bins=20, density=True, color='orange', edgecolor='black')
plt.plot(np.linspace(0, 1, 1000), beta.pdf(np.linspace(0, 1, 1000), a=1, b=1), 'g')
plt.title('Rozkład beta (1, 1)')


plt.subplot(2, 3, 5)
plt.hist(beta_samples2, bins=20, density=True, color='purple', edgecolor='black')
plt.plot(np.linspace(0, 1, 1000), beta.pdf(np.linspace(0, 1, 1000), a=2, b=2), 'g')
plt.title('Rozkład beta (2, 2)')


plt.subplot(2, 3, 6)
plt.hist(beta_samples3, bins=20, density=True, color='pink', edgecolor='black')
plt.plot(np.linspace(0, 1, 1000), beta.pdf(np.linspace(0, 1, 1000), a=3, b=9), 'g')
plt.title('Rozkład beta (3, 9)')

plt.tight_layout()
plt.show()
```
